```python
clean_df = df.dropna()
clean_df.tail()
```

```python
clean_df["Mid-Career 90th Percentile Salary"]-clean_df["Mid-Career 10th Percentile Salary"]
```

```python
Subtract = clean_df["Mid-Career 90th Percentile Salary"].subtract(clean_df["Mid-Career 10th Percentile Salary"])

clean_df.insert(1,"Subtract", Subtract)
clean_df.head()
```

```python
low_risk = clean_df.sort_values("Subtract")
low_risk[["Undergraduate Major", "Subtract"]].head()

```

```python
hp = clean_df.sort_values("Mid-Career 90th Percentile Salary", ascending=False)
hp[["Mid-Career 90th Percentile Salary","Undergraduate Major"]].head()
```

```python
hp = clean_df.sort_values("Subtract", ascending=False)
hp[["Mid-Career 90th Percentile Salary","Undergraduate Major","Subtract"]].head()
```

```python
hp = clean_df.sort_values("Mid-Career Median Salary", ascending=False)
hp[["Mid-Career Median Salary","Undergraduate Major","Subtract"]].head()
```
