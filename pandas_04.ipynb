```python
df_apps_clean[["App","Installs"]].groupby("Installs").count()
```

```python
df_apps_clean.Installs = df_apps_clean.Installs.astype(str).str.replace(",","")
df_apps_clean.Installs = pd.to_numeric(df_apps_clean.Installs)
df_apps_clean.loc[:,["App","Installs"]].groupby("Installs").nunique()
```

```python
df_apps_clean.Price.describe()
```

```python
df_apps_clean.Price = df_apps_clean.Price.astype(str).str.replace("$","")
df_apps_clean.Price = pd.to_numeric(df_apps_clean.Price)
df_apps_clean.sort_values("Price",ascending=False).head(20)
```

```python
df_apps_clean =  df_apps_clean[df_apps_clean["Price"] < 250]
df_apps_clean.sort_values("Price",ascending=False).head(5)
```

```python
df_apps_clean["Revenue_Estimate"] = df_apps_clean.Installs.mul(df_apps_clean.Price)
df_apps_clean.sort_values("Revenue_Estimate",ascending=False)[:10]
```

```python
top10_cat = df_apps_clean.Category.value_counts()[:10]
```

```python
bar = px.bar(x=top10_cat.index, title="Top 10 Category of App", y=top10_cat.values)
bar.show()
```

```python
cat_install = df_apps_clean.groupby("Category").agg({"Installs":pd.Series.sum})
cat_install.sort_values("Installs",ascending=True,inplace=True)
bar = px.bar(x=cat_install.Installs,y=cat_install.index, orientation="h",title="Top installed app category ")
bar.update_layout(xaxis_title="Number of Installs", yaxis_title="App Category")
bar.show()
```

```python
cat_app = df_apps_clean.groupby("Category").agg({"App":pd.Series.count})
```

```python
cat_install = df_apps_clean.groupby("Category").agg({"Installs":pd.Series.sum})
```

```python
cat_merge = pd.merge(cat_app, cat_install, on="Category", how="inner")
cat_merge.sort_values("Installs", ascending=False)
```

```python
scattar = px.scatter(cat_merge,x="App",y="Installs",size="App", title="Categorywise app Installs", color="Installs", hover_name=cat_merge.index)
scattar.update_layout(xaxis_title="Number of Apps", yaxis_title="Installs",yaxis = dict(type="log"))
scattar.show()
```

```python
stack = df_apps_clean.Genres.str.split(";",expand=True).stack()
num_gen = stack.value_counts()
```

```python
df_app_free_paid = df_apps_clean.groupby(["Category", "Type"],as_index=False).agg({"App":pd.Series.count})
```
